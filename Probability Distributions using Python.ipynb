### Normal Distribution

The Normal (or Gaussian) distribution is a continuous probability distribution that is often used to model real-world data. The Normal distribution is symmetrical around its mean and its spread is characterized by its standard deviation.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the mean and standard deviation of the normal distribution
mean = 0
std = 1

# Generate 100 random samples from a normal distribution
samples = np.random.normal(mean, std, 100)

# Plot the histogram of the samples to visualize the distribution
plt.hist(samples, bins=20, density=True)

# Add the normal distribution curve to the histogram
x = np.linspace(-5, 5, 100)
y = (1/(std * np.sqrt(2 * np.pi))) * np.exp(-(x-mean)**2/(2 * std**2))
plt.plot(x, y, 'r', linewidth=2)

plt.show()
```

### Binomial distribution

The Binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials, each with the same probability of success.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the Binomial distribution
n = 10  # number of trials
p = 0.5 # probability of success in each trial

# Generate 100 random samples from a Binomial distribution
samples = np.random.binomial(n, p, 100)

# Plot the histogram of the samples to visualize the distribution
plt.hist(samples, bins=20, density=True)

plt.show()
```

### Poisson distribution

The Poisson distribution is a discrete probability distribution that models the number of events that occur in a fixed interval of time or space.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameter for the Poisson distribution
lam = 3  # average number of events per interval

# Generate 100 random samples from a Poisson distribution
samples = np.random.poisson(lam, 100)

# Plot the histogram of the samples to visualize the distribution
plt.hist(samples, bins=20, density=True)

plt.show()
```

### Exponential Distribution

The Exponential distribution is a continuous probability distribution that models the time between events in a Poisson process. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameter for the Exponential distribution
lam = 1  # average rate of events

# Generate 100 random samples from an Exponential distribution
samples = np.random.exponential(1/lam, 100)

# Plot the histogram of the samples to visualize the distribution
plt.hist(samples, bins=20, density=True)

plt.show()
```

### Uniform Distribution

The Uniform distribution is a continuous probability distribution that models a situation where all values are equally likely within a specified range.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the Uniform distribution
lower = 0
upper = 10

# Generate 100 random samples from a Uniform distribution
samples = np.random.uniform(lower, upper, 100)

# Plot the histogram of the samples to visualize the distribution
plt.hist(samples, bins=20, density=True)

plt.show()
```
